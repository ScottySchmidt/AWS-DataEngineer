# AWS DataEngineer Pipeline

```python
import pandas as pd
import requests
from io import StringIO

series_id = "PRS85006091"  # ‚Üê choose any PR series from pr.series

url = "https://api.bls.gov/publicAPI/v2/timeseries/data/"

payload = {
    "seriesid": [series_id],
    "startyear": "1947",
    "endyear": "2030"
}

response = requests.post(url, json=payload)
data = response.json()

records = data["Results"]["series"][0]["data"]
df = pd.DataFrame(records)

df.head()
```
